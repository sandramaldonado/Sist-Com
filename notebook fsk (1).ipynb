```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
digital=[1,0,0,1,1,0,0,0,1,0,1,1]
```

```python
muestras=200
```

```python
muestreo=[]
```

```python
for i in range (1,12) :
  u=np.ones(muestras)
  m=u*digital[i]
  muestreo=np.concatenate((muestreo,m))
```

```python
plt.subplot(3,1,1)
plt.plot(muestreo)
```

```python
dimension=len(muestreo)
```

```python
t=np.linspace(0,5,dimension)
```

```python
frecs=5
frecm=25
```

```python
portadoras=np.cos(2*np.pi*frecs*t)
```

```python
portadoram=np.cos(2*np.pi*frecm*t)
```

```python
plt.subplot(4,1,2)
plt.plot(t,portadoras)
plt.subplot(4,1,3)
plt.plot(t,portadoram)
```

```python
modulada=[]
for i in range(0,dimension):
    if muestreo[i]==0:
         zero=np.array([portadoras[i]])
         modulada=np.concatenate((modulada,zero))
    else:
         uno=np.array([portadoram[i]])
         modulada=np.concatenate((modulada,uno))  
```

```python
plt.subplot(4,1,4)
plt.plot(t,modulada)
```

```python
plt.subplot(3,1,1)
plt.plot(muestreo)
plt.subplot(4,1,2)
plt.plot(t,portadoras)
plt.subplot(4,1,3)
plt.plot(t,portadoram)
plt.subplot(4,1,4)
plt.plot(t,modulada)
```

Explica en que consiste la modulación FSK
consiste basicamente en escoger entre dos señales portadoras de la misma amplitud pero con diferente frecuencia.

¿Por qué es necesario utilizar dos frecuencias en FSK?
porque la salida del FSK tiene q desplazarse entre las dos señales: la logica y la binaria. 

¿Qué relación existe entre la frecuencia de la portadora, la frecuencia espacio y la frecuencia marca?
son frecuencias sinuisoides pero la frec marca transmite un 1 y la frec espacio trasmite un 0
